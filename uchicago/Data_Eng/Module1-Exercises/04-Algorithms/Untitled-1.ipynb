```python
import math
from scipy.stats import poisson, binom

# Given parameters
lambda_per_km = 4.4  # flaws per km
lambda_per_m = lambda_per_km / 1000  # flaws per meter

# (A) Probability of exactly 1 flaw in 100m of cable using Poisson distribution
lambda_100m = lambda_per_m * 100  # Expected number of flaws in 100 meters
prob_1_flaw_100m = poisson.pmf(1, lambda_100m)

# (B) Probability that the distance between flaws exceeds 500m using Exponential distribution
lambda_exponential = lambda_per_m  # Rate for exponential distribution
prob_distance_exceeds_500m = math.exp(-lambda_exponential * 500)

# (C) Probability that at least 8 out of 10 segments are flaw-free using Binomial distribution
length_per_segment = 200  # meters
lambda_200m = lambda_per_m * length_per_segment  # Expected number of flaws in 200 meters
prob_no_flaw_200m = math.exp(-lambda_200m)  # P(0 flaws) in 200 meters

# Using binomial distribution to calculate P(at least 8 out of 10 are flaw-free)
n = 10  # number of segments
p = prob_no_flaw_200m  # probability of success (no flaws in 200m)

prob_at_least_8_free_of_flaws = binom.pmf(8, n, p) + binom.pmf(9, n, p) + binom.pmf(10, n, p)

# Results
prob_1_flaw_100m, prob_distance_exceeds_500m, prob_at_least_8_free_of_flaws
```
