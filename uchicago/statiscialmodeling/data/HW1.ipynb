```python
crashes = pd.read_csv('crashes.csv')
#fill na values with 0 on injuires fatal and incapacitating
crashes['INJURIES_FATAL'] = crashes['INJURIES_FATAL'].fillna(0)
crashes['INJURIES_INCAPACITATING'] = crashes['INJURIES_INCAPACITATING'].fillna(0)
crashes["SRINJURIES"]= crashes["INJURIES_FATAL"] + crashes["INJURIES_INCAPACITATING"]
#apply date format to CRASH_DATE
crashes['CRASH_DATE'] = pd.to_datetime(crashes['CRASH_DATE'])
```

```python
# Ensure 'CRASH_DATE' is the index for resampling
crashes.set_index('CRASH_DATE', inplace=True)

# Aggregate by the default weekly period (starting on Sunday), counting crashes and summing 'SRINJURIES'
weekly_crashes = crashes.resample('W').agg(
    Weekly_Crash_Count=('SRINJURIES', 'size'),     # Count of crashes
    Weekly_Severe_Injuries=('SRINJURIES', 'sum')   # Sum of severe injuries
)

# Reset the index if you want the date as a column
weekly_crashes = weekly_crashes.reset_index()

print(weekly_crashes)
```

```python
# Ensure 'CRASH_DATE' is the index for resampling
crashes.set_index('CRASH_DATE', inplace=True)

# Aggregate by week and count the number of crashes per week
weekly_crashes = crashes.resample('W').size()

# Rename the result column for clarity
weekly_crashes = weekly_crashes.rename('Weekly_Crash_Count')

# Convert back to a DataFrame if needed
weekly_crashes_df = weekly_crashes.reset_index()

print(weekly_crashes_df)
```

```python
# Group by week and sum the number of crashes and serious injuries
weekly_summary = crashes.groupby('WEEK').agg({
    'CRASH_DATE': 'count',   # Count of crashes per week
    'SRINJURIES': 'sum'      # Sum of serious injuries per week
}).rename(columns={'CRASH_DATE': 'NUM_CRASHES', 'SRINJURIES': 'TOTAL_SERIOUS_INJURIES'}).reset_index()
```
